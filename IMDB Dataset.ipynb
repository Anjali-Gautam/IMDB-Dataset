The Files provided are in .tsv format. In order to convert them to .csv and reading them in the python shell I have created a function in which I can input the name of the .tsv file and the name of the output .csv file I want

```python
# defining function for converting tsv to csv file and load it 
def file_gen(input_file,save_as):
    global df
    tsv_file= input_file
    csv_table=pd.read_table(tsv_file,sep='\t',low_memory=False)
    csv_table.to_csv(save_as,index=False)
    df = pd.read_csv(save_as,low_memory=True)
    return df
```

#### Merginng with Ratings table 

```python
### Ratings table
Rating_merge = pd.merge(movie_titles,
                       ratings_table[['tconst','averageRating','numVotes']],
                       left_on = 'titleId',
                       right_on = 'tconst',
                       how = 'inner')
Rating_merge.shape
#24528
```

```python
Rating_merge.head(15)
```

```python
Rating_merge.drop_duplicates(subset='titleId', keep='first', inplace=True)
Rating_merge.shape
```

#### Merging with title basics table

```python
#Merging with title basics 
Merge_table2 = pd.merge(Rating_merge,
                        title_basics,
                        left_on = 'titleId',
                        right_on = 'tconst',
                        how = 'left')
Merge_table2.shape
```

```python
Merge_table2.head(5)
```

##### Splitting the Generes 

```python
Merge_table2[['G1','G2','G3']]= Merge_table2.genres.str.split(",",expand=True)
```

drop tconst_x and tconst_y as the information is already in titleId

```python
Merge_table2.drop(['tconst_x','tconst_y'],axis=1)
```

#### Merging with title principals file

Title principals contains information of actors in a particular movie. Hence, it is expected that the number of rows/observations will increase after the merge

```python
#Merging with title principals
Merge_table3 = pd.merge(Merge_table2,
                        title_principals,
                        left_on = 'titleId',
                        right_on = 'tconst',
                        how = 'left')
Merge_table3.shape
```

```python
Merge_table3.head(5)
```

#### Merging with Name basics file

The name basics file contains the information on the actors in a movie and hence the merge will be on actor ID ( represented here by nconst).
Since the merge is on actor level the number of observations should remain same(when all actors are matched) or shoud4ld decrease (when some matches are not find) but will not increase

```python
#Merging with name basics 
Merge_table4 = pd.merge(Merge_table3,
                        name_basics,
                        left_on = 'nconst',
                        right_on = 'nconst',
                        how = 'left')
Merge_table4.shape
```
